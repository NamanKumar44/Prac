```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
df.shape
```

```python
df.columns = ["ID","sepallength","spealwidth","petallength","petalwidth","Species"]
```

```python
# since the values lies in different ranger we use scaler to scale them
#Min Max Scaler
```

```python
x = df.drop(['Species'],axis = 1)
y = df['Species']
```

```python
min(x['sepallength'])
```

```python
min(x['petallength'])
```

```python
max(x['petallength'])
```

```python
max(x['petalwidth'])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
x_scaled = scalar.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x_scaled,y,test_size = 0.5,random_state = 42)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train,Y_train)
```

```python
Y_pred = gnb.predict(X_test)
```

```python
Y_pred
```

```python
Y_test
```

```python
from sklearn import metrics
```

```python
import sklearn.metrics
lbs = ['Iris-versicolor','Iris-setosa','Iris-virginica']
print(sklearn.metrics.confusion_matrix(Y_test,Y_pred,labels = lbs))
print(sklearn.metrics.accuracy_score(Y_test,Y_pred))
print(sklearn.metrics.precision_score(Y_test,Y_pred,labels = lbs,pos_label = 1,average = 'micro'))
print(sklearn.metrics.f1_score(Y_test,Y_pred,labels = lbs,pos_label = 1,average = 'micro'))
print(sklearn.metrics.recall_score(Y_test,Y_pred,labels = lbs,pos_label = 1,average = 'micro'))
```
